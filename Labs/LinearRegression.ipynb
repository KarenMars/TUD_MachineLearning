## Polynomial Regression and Other Feature Transformations
Polynomial regression fits a polinomial of some maximal degree of the data in a least squaure sense.
Even though this results in a nonlinear function in x, the problem may still be preferred to as linear regression as the regression function. Linear in the unknown parameters w estimated from the data. 

```python
import sys
sys.path.append('/Users/jianlan/Desktop/prtools')
import prtools as pr
import numpy as np
import matplotlib.pyplot as plt
```

### Exercise 2.5

```python
# generate data where inputs are drawn uniformly from [0,1]
# study the polynomials of different degree 0 to 3 for different training sizes (e.g. 4,40,400 samples)
# solving linear equations
n = 4
x = np.random.rand(n,1)
y = x*x + np.random.randn(n,1)
a = pr.gendatr(x,y)
w = pr.linearr(a)
w_2 = pr.linearr(a,2)
w3 = pr.linearr(a,3)
pr.scatterr(a)
pr.plotr(w,color='red')
pr.plotr(w_2,color='green')
pr.plotr(w3,color='blue')
```
